<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">1.3.1 - Nettoyage du fichier aide alimentaire</h3>

```python
#Renommination du champ Pays beneficiaire en Zone
aide_alim = aide_alim[(aide_alim['Année'] == 2016)].groupby("Pays bénéficiaire").sum().reset_index()
aide_alim.rename(columns={"Pays bénéficiaire":"Zone","Valeur":"Aide alimentaire"}, inplace=True)
aide_alim.head()
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">1.3.2 - Nettoyage du fichier disponibilité alimentaire</h3>
</div>

```python
dispo_alim2017['Élément'].unique()
```

```python
dispo_alim2017['Produit'].unique()
```

```python
df_volaille=dispo_alim2017[dispo_alim2017['Produit']=='Viande de Volailles']
df_volaille.head()
```

```python
df_pivot = df_volaille.pivot(index=['Zone','Année'],columns=['Élément'],values='Valeur').fillna(0).reset_index()
df_pivot.head()
```

```python
# Liste des colonnes à convertir
a_convertir = ['Disponibilité intérieure', 'Exportations - Quantité', 'Importations - Quantité', 'Nourriture', 'Production']

# Vérifier que les colonnes existent dans le DataFrame
colonnes_existantes = [c for c in a_convertir if c in df_pivot.columns]

# Appliquer la conversion uniquement aux colonnes existantes et numériques
for c in colonnes_existantes:
    if pd.api.types.is_numeric_dtype(df_pivot[c]):  # Vérifier que la colonne est numérique
        df_pivot[c] = df_pivot[c].fillna(0) * 1_000_000  # Remplacer NaN par 0 avant la multiplication
    else:
        print(f"Avertissement : La colonne '{c}' n'est pas numérique et n'a pas été convertie.")
```

```python
Colonne_garder=['Zone','Disponibilité intérieure', 'Exportations - Quantité', 'Importations - Quantité', 'Nourriture', 'Production']
df_aide_alimnet=df_pivot[Colonne_garder]
df_aide_alimnet.head()
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">1.3.3 - Nettoyage du fichier Population</h3>
</div>

```python
Population = Population[Population['Année'].isin([2017,2018])]
Population.head()
```

```python
Population['Année'].unique()
```

```python
df_pivot_pop = Population.pivot(index=['Zone'], columns=['Année'], values='Valeur').fillna(0).reset_index()

# Calcul de la variation de population entre 2017 et 2018
df_pivot_pop['Variation_population'] = round((df_pivot_pop[2018] - df_pivot_pop[2017]) / df_pivot_pop[2017] * 100, 2)

# Suppression des colonnes 2017 et 2018
df_pivot_pop.drop(columns=[2017, 2018], inplace=True)
df_pivot_pop
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">1.3.4 - Nettoyage du fichier PIB</h3>
</div>

```python
df_pib2017=pib[pib['Année']==2017]
colonnes_garderpib=['Zone','Valeur']
pib_2017 = df_pib2017[colonnes_garderpib]
pib_2017.rename(columns={"Valeur": "PIB"}, inplace=True)
```

```python
pib_2017
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">2 - Jointure des fichiers </h3>
</div>

```python
# Fusion des DataFrames
df_final = df_pivot_pop.merge(df_aide_alimnet, on= "Zone", how="left")\
    .merge(aide_alim, on= "Zone", how="left")\
    .merge(pib_2017, on= "Zone", how="left")
df_final.head()
```

```python
df_final['Produit'].unique()
```

```python
#Remplacement des valeurs manquantes par 0
df_final.fillna(0, inplace=True)
df_final
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3 - Export du fichier global  </h3>

```python
Colonnes_agarderfinal=['Zone','Variation_population', 'Disponibilité intérieure', 'Exportations - Quantité', 'Importations - Quantité', 'Nourriture','Production','Aide alimentaire','PIB']
df_final=df_final[Colonnes_agarderfinal]
df_final.head()
```

```python
# Exporter vers un fichier Excel
fichier_global = "export_df_final.xlsx"
# Exporter vers Excel
df_final.to_excel(fichier_global, index=False)
print(f"Fichier Excel enregistré sous : {fichier_global}")
```
